```python
#==> ENTER YOUR CODE HERE
# What is the average tip for trips paid for with credit card?
print(f' - The average tip for trips paid for with a credit card is: {df[df["payment_type"] == 1]["tip_amount"].mean()}')


#==> ENTER YOUR CODE HERE
# What is the average tip for trips paid for with cash ?
print(f' - The average tip for trips paid for with cash is: {df[df["payment_type"] == 2]["tip_amount"].mean()}')
```

```python
#==> ENTER YOUR CODE HERE

# How many times is each vendor ID represented in the data?
df["VendorID"].value_counts()
```

```python
#==> ENTER YOUR CODE HERE

# What is the mean total amount for each vendor?
df.groupby("VendorID")["total_amount"].mean()
```

```python
#==> ENTER YOUR CODE HERE

# Filter the data for credit card payments only
df_credit_card_payments = df[df["payment_type"] == 1]

#==> ENTER YOUR CODE HERE

# Filter the credit-card-only data for passenger count only
average_tip_by_passenger_count = df_credit_card_payments.groupby("passenger_count", as_index=False)["tip_amount"].mean()
```

```python
#==> ENTER YOUR CODE HERE

# Calculate the average tip amount for each passenger count (credit card payments only)
average_tip_by_passenger_count.rename(columns={"tip_amount": "average_tip_amount"}, inplace=True)
average_tip_by_passenger_count
```

```python
plt.figure(figsize=(10, 6)) 
sns.barplot(x='passenger_count', y='average_tip_amount', data=average_tip_by_passenger_count, palette='viridis')

plt.xlabel('Passenger Count')
plt.ylabel('Average Tip Amount ($)')
plt.title('Average Tip Amount by Passenger Count for Credit Card Payments')

plt.show()
```
